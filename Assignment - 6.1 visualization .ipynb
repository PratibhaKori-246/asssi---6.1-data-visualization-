Q 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Prepare the output for notebook
output_notebook()

# Step 2: Generate a dataset with categories and numerical values
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data_points_per_category = 50  # Number of data points per category

# Generate random numerical values for each category
data = {
    'Category': np.repeat(categories, data_points_per_category),
    'Value': np.random.randn(data_points_per_category * len(categories)) * 10 + 50  # Normal distribution
}

# Step 3: Create a DataFrame
df = pd.DataFrame(data)

# Step 4: Create a box plot to visualize the distribution of values across categories
p = figure(title="Distribution of Values Across Categories", 
           x_axis_label='Category', 
           y_axis_label='Value', 
           width=600, 
           height=400)

# Step 5: Create a color palette
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Step 6: Add box glyphs for each category
for i, category in enumerate(categories):
    p.quad(top=df[df['Category'] == category]['Value'].quantile(0.75), 
            bottom=df[df['Category'] == category]['Value'].quantile(0.25), 
            left=i - 0.2, right=i + 0.2, 
            fill_color=colors[i], line_color='black', 
            legend_label=category, alpha=0.6)

# Step 7: Add a legend
p.legend.title = "Categories"

# Step 8: Display the plot
show(p)
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
n_samples = 100  # Number of samples

# Generate correlated features
feature_1 = np.random.randn(n_samples)
feature_2 = feature_1 * 0.8 + np.random.randn(n_samples) * 0.2  # Strong correlation with feature_1
feature_3 = feature_1 * 0.2 + np.random.randn(n_samples) * 0.5  # Weak correlation with feature_1
feature_4 = np.random.randn(n_samples)  # No correlation

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Compute the correlation matrix
correlation_matrix = df.corr()    # Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()  # Show color scale

# Set ticks and labels
plt.xticks(ticks=np.arange(len(correlation_matrix)), labels=correlation_matrix.columns, rotation=45)
plt.yticks(ticks=np.arange(len(correlation_matrix)), labels=correlation_matrix.columns)

# Add title
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
plt.show()                                                                                                                                                                                                                                
```

2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a synthetic dataset for student grades
np.random.seed(15)
data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data_grades)

# Step 2: Create a violin-like plot (boxplot with jittered points)
plt.figure(figsize=(10, 6))

# Create a boxplot for each grade category
for i, grade in enumerate(sorted(df_grades['Grade'].unique())):
    scores = df_grades[df_grades['Grade'] == grade]['Score']
    # Plot boxplot
    plt.boxplot(scores, positions=[i], widths=0.6, patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'))
    # Add jittered points
    jitter = np.random.normal(0, 0.04, size=len(scores))
    plt.scatter(np.full(scores.shape, i) + jitter, scores, alpha=0.5)

plt.title('Distribution of Scores Across Different Grade Categories')
plt.xticks(ticks=np.arange(len(sorted(df_grades['Grade'].unique())), dtype=int), labels=sorted(df_grades['Grade'].unique()))
plt.xlabel('Grade')
plt.ylabel('Score')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = x**2 + y**2  # Simple function

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(x, y, z, cmap='coolwarm')

# Add color bar, labels, and title
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'$z = x^2 + y^2$')

plt.show()
```

5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)


```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.io import show
from bokeh.palettes import Viridis256
import numpy as np

# Step 1: Prepare the output for notebook
output_notebook()

# Step 2: Create the dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Step 3: Prepare the data for the heatmap
heatmap_data = data_heatmap.flatten()
x_coords = xx.flatten()
y_coords = yy.flatten()

# Step 4: Create a ColumnDataSource
source = ColumnDataSource(data=dict(value=heatmap_data, x=x_coords, y=y_coords))

# Step 5: Create a figure for the heatmap
p = figure(title="Heatmap", 
           x_axis_label='X', 
           y_axis_label='Y', 
           width=600, 
           height=600, 
           tools="")


# Step 6: Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(heatmap_data), high=max(heatmap_data))

# Step 7: Add squares to the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        line_color=None, fill_color=mapper)

# Step 8: Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

# Step 9: Display the heatmap
show(p)
```
